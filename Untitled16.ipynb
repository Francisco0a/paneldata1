```python
"""

Project Econometrics JBC

"""

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from linearmodels.panel import PanelOLS, RandomEffects, FirstDifferenceOLS
import statsmodels.api as sm
from google.colab import drive
drive.mount('/content/drive')
pd.set_option('display.max_columns', None)
# Uploading the filtered dataset with the department code 69 as a parquet file
####1.Transform if necessary the database in the right format.
####2.	Select your “département.” in the code.

#%% Section 1
# filtered_df = diane_data[diane_data['départementcode'] == 69]

file_path = '/content/drive/My Drive/filtered_df.parquet'
filtered_df = pd.read_parquet(file_path)

# Selection of Department
print(filtered_df.columns)

filtered_data = filtered_df[filtered_df['départementcode'] == 69.0]
print(filtered_data.head())

# Computing the ratio
# (CASH+MARKETABLE SECURITIES)/TOTAL ASSETS
# Step 1: Compute the ratio where conditions are met
filtered_data['ratio'] = np.where(
    (filtered_data['cashNbank'] + filtered_data['marketable_sec'] >= 0) & (filtered_data['totalassets'] >= 1),
    (filtered_data['cashNbank'] + filtered_data['marketable_sec']) / filtered_data['totalassets'],
    np.nan
)

# Step 2: For each firm, calculate the lag of the computed ratio
filtered_data['ratio_lag'] = filtered_data.groupby('siren')['ratio'].shift(1)

# Check that the first data point for each firm is missing
missing_first_data = filtered_data.groupby('siren').apply(lambda x: pd.isna(x['ratio_lag'].iloc[0])).all()
if missing_first_data:
    print("For each firm, the first data point of the lagged ratio is missing.")
else:
    print("For at least one firm, the first data point of the lagged ratio is not missing.")

numeric_variables = ['cashNbank', 'marketable_sec', 'totalassets', 'netsales', 'netincome', 'totalexpenses', 'inventory', 'creditors', 'totalDebt', 'totalfixedassets']
for var in numeric_variables:
    filtered_data[var] = pd.to_numeric(filtered_data[var], errors='coerce')

# Compute the trasformed variables
filtered_data['CASH'] = (filtered_data['cashNbank'] + filtered_data['marketable_sec']) / filtered_data['totalassets']
filtered_data['GROWTH'] = filtered_data.sort_values(by=['year']).groupby('siren')['netsales'].pct_change()
filtered_data['SIZE'] = np.log(filtered_data['totalassets'])
filtered_data['CFLOW'] = (filtered_data['netincome'] + (filtered_data['totalexpenses'] - filtered_data['netsales'])) / filtered_data['totalassets']
filtered_data['NWC'] = ((filtered_data['inventory'] + filtered_data['creditors']) - (filtered_data['cashNbank'] + filtered_data['marketable_sec'])) / filtered_data['totalassets']
filtered_data['LEV'] = filtered_data['totalDebt'] / filtered_data['totalassets']
filtered_data['INCASSET'] = filtered_data.groupby('siren')['totalfixedassets'].pct_change() / filtered_data['totalassets']

#filtered_data['year'] = pd.to_datetime(filtered_data['year'], errors='coerce').dt.year
#filtered_data['siren'] = pd.to_numeric(filtered_data['siren'], errors='coerce')

variables = ['CASH', 'GROWTH', 'SIZE', 'CFLOW', 'NWC', 'LEV', 'INCASSET']

for var in variables:
    # Determine the 1st and 99th percentiles
    lower_bound = filtered_data[var].quantile(0.01)
    upper_bound = filtered_data[var].quantile(0.99)

    # Cap outliers at the 1st and 99th percentile values
    filtered_data[var] = np.where(filtered_data[var] < lower_bound, lower_bound, filtered_data[var])
    filtered_data[var] = np.where(filtered_data[var] > upper_bound, upper_bound, filtered_data[var])

# Initialize a DataFrame to store transformed variables
transformed_df = pd.DataFrame(index=filtered_data.index)

# Compute transformations
for var in variables:
    # Between transformation: Mean for each 'siren' (entity)
    transformed_df[f'{var}_between'] = filtered_data.groupby('siren')[var].transform('mean')

    # One-way within transformation: Value - Mean for each 'siren'
    transformed_df[f'{var}_one_way_within'] = filtered_data[var] - transformed_df[f'{var}_between']

    # First differences transformation: Current period value - Previous period value for each 'siren'
    transformed_df[f'{var}_first_diff'] = filtered_data.groupby('siren')[var].diff()

grand_means = filtered_data[variables].mean()

# Calculate the mean for each entity (siren) across time
entity_means = filtered_data.groupby('siren')[variables].transform('mean')

# Calculate the mean for each time period across all entities
time_means = filtered_data.groupby('year')[variables].transform('mean')

# Initialize a DataFrame to store the two-way within transformed variables
two_way_within_transformed = pd.DataFrame(index=filtered_data.index)

# Perform the two-way within transformation
for var in variables:
    two_way_within_transformed[f'{var}_two_way_within'] = (
        filtered_data[var] -
        entity_means[var] -
        time_means[var] +
        grand_means[var]
    )

#exercise 5
unique_years = sorted(filtered_data['year'].unique())

# Print the unique years
print(unique_years)

grouped = filtered_data.groupby('year')['CASH'].agg(['count', 'mean', 'std', 'median', lambda x: x.quantile(0.1), lambda x: x.quantile(0.9)])
grouped.columns = ['Obs', 'Mean', 'Std. dev.', 'Median', 'Perc. 10', 'Perc. 90']

# Calculate aggregate statistics for the entire dataset
aggregate_stats = filtered_data['CASH'].agg(['count', 'mean', 'std', 'median', lambda x: x.quantile(0.1), lambda x: x.quantile(0.9)]).to_frame().T
aggregate_stats.index = ['1998-2012']
aggregate_stats.columns = ['Obs', 'Mean', 'Std. dev.', 'Median', 'Perc. 10', 'Perc. 90']

# Concatenate yearly statistics with aggregate statistics
final_table = pd.concat([grouped, aggregate_stats])

# Formatting 'Obs' to include commas for thousands
final_table['Obs'] = final_table['Obs'].apply(lambda x: f"{x:,.0f}")

print(final_table)

#exercise 6
# First, ensure 'CASH' is numeric if not already
filtered_data['CASH'] = pd.to_numeric(filtered_data['CASH'], errors='coerce')

# Within Transformation
filtered_data['CASH_within'] = filtered_data.groupby('siren')['CASH'].transform(lambda x: x - x.mean())

# First Differences
filtered_data['CASH_first_diff'] = filtered_data.groupby('siren')['CASH'].diff()

# Verify transformations
print(filtered_data[['CASH_within', 'CASH_first_diff']].head())

# Perform groupby for within-transformed 'CASH' and calculate statistics
grouped_within = filtered_data.groupby('year')['CASH_within'].agg(['count', 'mean', 'std', 'median', lambda x: x.quantile(0.1), lambda x: x.quantile(0.9)])
grouped_within.columns = ['Obs', 'Mean', 'Std. dev.', 'Median', 'Perc. 10', 'Perc. 90']

# Calculate aggregate statistics for within-transformed 'CASH'
aggregate_stats_within = filtered_data['CASH_within'].agg(['count', 'mean', 'std', 'median', lambda x: x.quantile(0.1), lambda x: x.quantile(0.9)]).to_frame().T
aggregate_stats_within.index = ['Aggregate']
aggregate_stats_within.columns = ['Obs', 'Mean', 'Std. dev.', 'Median', 'Perc. 10', 'Perc. 90']

# Perform groupby for first differences 'CASH' and calculate statistics
grouped_first_diff = filtered_data.groupby('year')['CASH_first_diff'].agg(['count', 'mean', 'std', 'median', lambda x: x.quantile(0.1), lambda x: x.quantile(0.9)])
grouped_first_diff.columns = ['Obs', 'Mean', 'Std. dev.', 'Median', 'Perc. 10', 'Perc. 90']

# Calculate aggregate statistics for first differences 'CASH'
aggregate_stats_first_diff = filtered_data['CASH_first_diff'].agg(['count', 'mean', 'std', 'median', lambda x: x.quantile(0.1), lambda x: x.quantile(0.9)]).to_frame().T
aggregate_stats_first_diff.index = ['Aggregate']
aggregate_stats_first_diff.columns = ['Obs', 'Mean', 'Std. dev.', 'Median', 'Perc. 10', 'Perc. 90']


# Concatenate yearly statistics with aggregate statistics for both transformations
final_table_within = pd.concat([grouped_within, aggregate_stats_within])
final_table_first_diff = pd.concat([grouped_first_diff, aggregate_stats_first_diff])

# Formatting 'Obs' to include commas for thousands for both tables
final_table_within['Obs'] = final_table_within['Obs'].apply(lambda x: f"{x:,.0f}")
final_table_first_diff['Obs'] = final_table_first_diff['Obs'].apply(lambda x: f"{x:,.0f}")

# Print the tables
print("Within Transformed Cash Holdings:")
print(final_table_within)
print("\nFirst Differences in Cash Holdings:")
print(final_table_first_diff)

#exercise 7
variables = ['CASH', 'GROWTH', 'SIZE', 'CFLOW', 'NWC', 'LEV','INCASSET']

# Calculate descriptive statistics for each variable
descriptive_stats = filtered_data[variables].agg(['count', 'mean', 'std', 'median', lambda x: x.quantile(0.1), lambda x: x.quantile(0.9)]).T
descriptive_stats.columns = ['Obs', 'Mean', 'Std. Dev.', 'Median', 'Perc. 10', 'Perc. 90']

# Adjust 'Obs' format to include commas as thousands separators
descriptive_stats['Obs'] = descriptive_stats['Obs'].apply(lambda x: f"{x:,.0f}")

# Print the table
print("Table 1 Descriptive Statistics")
print(descriptive_stats)

#exercise 8
transformed_data = pd.DataFrame()

# Assuming 'year' and 'siren' are in the right format
variables = ['CASH', 'GROWTH', 'SIZE', 'CFLOW', 'NWC', 'LEV', 'INCASSET']

# Between transformation
for var in variables:
    transformed_data[f'{var}_between'] = filtered_data.groupby('siren')[var].transform('mean')

# One-way within transformation
for var in variables:
    group_means = filtered_data.groupby('siren')[var].transform('mean')
    transformed_data[f'{var}_one_way_within'] = filtered_data[var] - group_means

# Two-way within transformation (adjust for specific needs)
for var in variables:
    entity_means = filtered_data.groupby('siren')[var].transform('mean')
    time_means = filtered_data.groupby('year')[var].transform('mean')
    overall_mean = filtered_data[var].mean()
    transformed_data[f'{var}_two_way_within'] = filtered_data[var] - entity_means - time_means + overall_mean

# First differences
for var in variables:
    transformed_data[f'{var}_first_diff'] = filtered_data.groupby('siren')[var].diff()

# Function to calculate descriptive statistics
def calculate_descriptive_stats(df, variable_suffix):
    descriptive_stats = df.filter(like=variable_suffix).agg(['count', 'mean', 'std', 'median', lambda x: x.quantile(0.1), lambda x: x.quantile(0.9)])
    descriptive_stats = descriptive_stats.T
    descriptive_stats.columns = ['Obs', 'Mean', 'Std. Dev.', 'Median', 'Perc. 10', 'Perc. 90']
    descriptive_stats['Obs'] = descriptive_stats['Obs'].apply(lambda x: f"{x:,.0f}")
    return descriptive_stats

# Calculate for each transformation
stats_between = calculate_descriptive_stats(transformed_data, 'between')
stats_one_way_within = calculate_descriptive_stats(transformed_data, 'one_way_within')
stats_two_way_within = calculate_descriptive_stats(transformed_data, 'two_way_within')
stats_first_diff = calculate_descriptive_stats(transformed_data, 'first_diff')


#exercise 9
var_cash_original = filtered_data['CASH'].var()

# Within Transformation: CASH - Mean(CASH) per siren
cash_within = filtered_data['CASH'] - filtered_data.groupby('siren')['CASH'].transform('mean')
var_cash_within = cash_within.var()

# Between Transformation: Mean(CASH) per siren
cash_between = filtered_data.groupby('siren')['CASH'].transform('mean')
var_cash_between = cash_between.var()

# Proportion of variance
total_var = var_cash_within + var_cash_between
prop_within = var_cash_within / total_var
prop_between = var_cash_between / total_var

# Print Variances and Proportions
print(f"Variance of Original CASH: {var_cash_original}")
print(f"Variance of Within-Transformed CASH: {var_cash_within}")
print(f"Variance of Between-Transformed CASH: {var_cash_between}")
print(f"Proportion of Variance Explained by Within Transformation: {prop_within}")
print(f"Proportion of Variance Explained by Between Transformation: {prop_between}")
# Print tables
print("Table 1B: Between Transformed Variables")
print(stats_between)
print("\nTable 1W: One-Way Within Transformed Variables")
print(stats_one_way_within)
print("\nTable 2W: Two-Way Within Transformed Variables")
print(stats_two_way_within)
print("\nTable 1D: First Differences")
print(stats_first_diff)

#exercise 10
variables = ['CASH', 'GROWTH', 'SIZE', 'CFLOW', 'NWC', 'LEV', 'INCASSET']

# Between transformation
between_transformed = filtered_data.groupby('siren')[variables].transform('mean')

# One-way within transformation
one_way_within_transformed = filtered_data[variables].apply(lambda x: x - filtered_data.groupby('siren')[x.name].transform('mean'))

corr_matrix_original = filtered_data[variables].corr()
print(corr_matrix_original)

# Between-transformed correlation matrix
corr_matrix_between = between_transformed.corr()
print(corr_matrix_between)

# One-way within-transformed correlation matrix
corr_matrix_one_way_within = one_way_within_transformed.corr()
print(corr_matrix_one_way_within)

# Set the style of the visualization
sns.set(style="white")

# Define a function to create a heatmap
def plot_heatmap(corr_matrix, title):
    # Generate a mask for the upper triangle
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

    # Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(11, 9))

    # Generate a custom diverging colormap
    cmap = sns.diverging_palette(230, 20, as_cmap=True)

    # Draw the heatmap with the mask and correct aspect ratio
    sns.heatmap(corr_matrix, mask=mask, cmap=cmap, vmax=.3, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

    plt.title(title)

# Plot heatmaps
plot_heatmap(corr_matrix_original, 'Correlation Matrix (Original)')
plot_heatmap(corr_matrix_between, 'Correlation Matrix (Between Transformed)')
plot_heatmap(corr_matrix_one_way_within, 'Correlation Matrix (One-Way Within Transformed)')

plt.show()

#%% exercice 11
from statsmodels.regression.linear_model import OLS
from linearmodels import BetweenOLS, PanelOLS, RandomEffects, FirstDifferenceOLS
from statsmodels.tools.tools import add_constant

# Créer un nouveau DataFrame pour stocker les données filtrées
filtered_data_new = filtered_data.copy()

# Créer une colonne pour CASH_lag
filtered_data_new['CASH_lag'] = filtered_data.groupby('siren')['CASH'].shift(1)

# Supprimer les lignes avec des valeurs manquantes uniquement dans la colonne 'CASH_lag'
filtered_data_new = filtered_data_new.dropna(subset=['CASH_lag','CASH'])

filtered_data_new.set_index(['siren', 'year'], inplace=True)

# Préparation des variables explicatives (X) et dépendantes (Y)
X = add_constant(filtered_data_new['CASH_lag'])
Y = filtered_data_new['CASH']

# OLS
model_ols = OLS(Y, X)
results_ols = model_ols.fit()
print("OLS Results:\n", results_ols.summary())

# Between
model_between = BetweenOLS(Y, X)
results_between = model_between.fit()
print("\nBetween Results:\n", results_between)

# Within (Fixed Effects)
model_within = PanelOLS(Y, X, entity_effects=True)
results_within = model_within.fit()
print("\nWithin Results:\n", results_within)

# Random Effects
model_re = RandomEffects(Y, X)
results_re = model_re.fit()
print("\nRandom Effects Results:\n", results_re)

# First Differences
# Pour FirstDifferenceOLS, enlever 'const' car cette méthode n'utilise pas de constante
X_fd = X.drop(columns='const')
model_fd = FirstDifferenceOLS(Y, X_fd)
results_fd = model_fd.fit()
print("\nFirst Differences Results:\n", results_fd)


#%% exercise 11
"""
from statsmodels.regression.linear_model import OLS
from linearmodels import BetweenOLS, PanelOLS, RandomEffects, FirstDifferenceOLS
from statsmodels.tools.tools import add_constant

# Create lagged CASH variable
filtered_data['CASH_lag'] = filtered_data.groupby('siren')['CASH'].shift(1)

# Drop NA values that result from lagging
filtered_data = filtered_data.dropna()

# For fixed effects model, set index
panel_data = filtered_data.set_index(['siren', 'year'])

# Préparation des variables explicatives (X) et dépendantes (Y)
X = panel_data['CASH_lag']
X = add_constant(X)
Y = panel_data['CASH']

# OLS
model_ols = OLS(Y, X)
results_ols = model_ols.fit()
print("OLS Results:\n", results_ols.summary())

# Between
model_between = BetweenOLS(Y, X)
results_between = model_between.fit()
print("\nBetween Results:\n", results_between)

# Within (Fixed Effects)
model_within = PanelOLS(Y, X, entity_effects=True)
results_within = model_within.fit()
print("\nWithin Results:\n", results_within)

# Random Effects
model_re = RandomEffects(Y, X)
results_re = model_re.fit()
print("\nRandom Effects Results:\n", results_re)

# First Differences
# Pour FirstDifferenceOLS, enlever 'const' car cette méthode n'utilise pas de constante
X_fd = X.drop(columns='const')
model_fd = FirstDifferenceOLS(Y, X_fd)
results_fd = model_fd.fit()
print("\nFirst Differences Results:\n", results_fd)

"""

#%% exercice 12
# cleaning data
panel_data = filtered_data_new.copy()
panel_data.dropna(subset=['CASH_lag', 'GROWTH', 'SIZE', 'CFLOW', 'NWC', 'LEV', 'INCASSET'], inplace=True)

# Préparation des variables explicatives (X) et dépendantes (Y) incluant les variables de contrôle
X_vars = ['CASH_lag', 'GROWTH', 'SIZE', 'CFLOW', 'NWC', 'LEV', 'INCASSET']
X = panel_data[X_vars]
X = add_constant(X)
Y = panel_data['CASH']

# OLS
model_ols = OLS(Y, X).fit()
print("OLS Results:\n", model_ols.summary())

# Between
model_between = BetweenOLS(Y, X).fit()
print("\nBetween Results:\n", model_between.summary)

# Within (Fixed Effects)
model_within = PanelOLS.from_formula('CASH ~ CASH_lag + GROWTH + SIZE + CFLOW + NWC + LEV + INCASSET + EntityEffects', data=panel_data).fit()
print("\nWithin Results:\n", model_within.summary)

#%% exercice 13
# Créer une copie du DataFrame filtered_data
filtered_data_copy = filtered_data_new.copy()

# Calculate the increase in fixed assets
filtered_data_copy['INCASSET_DIFF'] = filtered_data_copy.groupby('siren')['INCASSET'].diff()

# Calculate the median increase in fixed assets across the sample
median_increase = filtered_data_copy['INCASSET_DIFF'].median()

# Create the dummy variable
filtered_data_copy['DHIGHINCASSET'] = (filtered_data_copy['INCASSET_DIFF'] > median_increase).astype(int)

# Create the interaction term between lagged CASH and the dummy variable
filtered_data_copy['CASH_lag_X_DHIGHINCASSET'] = filtered_data_copy['CASH_lag'] * filtered_data_copy['DHIGHINCASSET']

# Drop NA values that might result from the lagging and differencing operations
filtered_data_copy.dropna(subset=['CASH','CASH_lag', 'GROWTH', 'SIZE', 'CFLOW', 'NWC', 'LEV', 'INCASSET', 'DHIGHINCASSET', 'CASH_lag_X_DHIGHINCASSET'], inplace=True)

# X and Y
X_vars = ['CASH_lag', 'DHIGHINCASSET', 'CASH_lag_X_DHIGHINCASSET', 'GROWTH', 'SIZE', 'CFLOW', 'NWC', 'LEV', 'INCASSET']
X = add_constant(filtered_data_copy[X_vars])
Y = filtered_data_copy['CASH']

# OLS
model_ols = OLS(Y, X).fit()
print("OLS Results:\n", model_ols.summary())

# Between
model_between = BetweenOLS(Y, X).fit()
print("\nBetween Results:\n", model_between.summary)

# Within (Fixed Effects)
model_within = PanelOLS.from_formula('CASH ~ CASH_lag + DHIGHINCASSET + CASH_lag_X_DHIGHINCASSET + GROWTH + SIZE + CFLOW + NWC + LEV + INCASSET + EntityEffects', data=filtered_data_copy).fit()
print("\nWithin Results:\n", model_within.summary)

# For FirstDifferenceOLS, remove 'const' as this method does not use a constant
X_fd = X.drop(columns='const')
model_fd = FirstDifferenceOLS(Y, X_fd).fit()
print("\nFirst Differences Results:\n", model_fd.summary)

"""
#%% exercice 13
# We want to replicate column (2)
# Calculate the Dummy Variable and Interaction Term
# Calculate the increase in fixed assets
filtered_data['INCASSET_DIFF'] = filtered_data.groupby('siren')['INCASSET'].diff()

# Calculate the median increase in fixed assets across the sample
median_increase = filtered_data['INCASSET_DIFF'].median()

# Create the dummy variable
filtered_data['DHIGHINCASSET'] = (filtered_data['INCASSET_DIFF'] > median_increase).astype(int)

# Create the interaction term between lagged CASH and the dummy variable
filtered_data['CASH_lag_X_DHIGHINCASSET'] = filtered_data['CASH_lag'] * filtered_data['DHIGHINCASSET']

# Drop NA values that might result from the lagging and differencing operations
filtered_data.dropna(subset=['CASH_lag', 'GROWTH', 'SIZE', 'CFLOW', 'NWC', 'LEV', 'INCASSET', 'DHIGHINCASSET', 'CASH_lag_X_DHIGHINCASSET'], inplace=True)

# For fixed effects model, set index
panel_data = filtered_data.set_index(['siren', 'year'])

# X and Y
X_vars = ['CASH_lag', 'DHIGHINCASSET', 'CASH_lag_X_DHIGHINCASSET', 'GROWTH', 'SIZE', 'CFLOW', 'NWC', 'LEV', 'INCASSET']
X = panel_data[X_vars]
X = add_constant(X)
Y = panel_data['CASH']

# OLS
model_ols = OLS(Y, X).fit()
print("OLS Results:\n", model_ols.summary())

# Between
model_between = BetweenOLS(Y, X).fit()
print("\nBetween Results:\n", model_between.summary)

# Within (Fixed Effects)
model_within = PanelOLS.from_formula('CASH ~ CASH_lag + DHIGHINCASSET + CASH_lag_X_DHIGHINCASSET + GROWTH + SIZE + CFLOW + NWC + LEV + INCASSET + EntityEffects', data=panel_data).fit()
print("\nWithin Results:\n", model_within.summary)

# For FirstDifferenceOLS, remove 'const' as this method does not use a constant
X_fd = X.drop(columns='const')
model_fd = FirstDifferenceOLS(Y, X_fd).fit()
print("\nFirst Differences Results:\n", model_fd.summary)
"""
#%%
# Calculer les lags et les premières différences nécessaires
filtered_data['CASH_lag1'] = filtered_data.groupby('siren')['CASH'].shift(1)
filtered_data['delta_CASH'] = filtered_data['CASH'].diff()

control_vars = ['SIZE', 'GROWTH', 'CFLOW', 'NWC', 'LEV', 'INCASSET']

for var in control_vars:
    filtered_data[f'{var}_lag1'] = filtered_data.groupby('siren')[var].shift(1)
    filtered_data[f'delta_{var}'] = filtered_data[var].diff()

# Déterminer toutes les colonnes qui pourraient avoir des NaN après le lagging et differencing
cols_with_potential_nans = ['CASH_lag1', 'delta_CASH'] + [f'{var}_lag1' for var in control_vars] + [f'delta_{var}' for var in control_vars]

# Supprimer les observations avec des valeurs manquantes dans ces colonnes spécifiques
filtered_data.dropna(subset=cols_with_potential_nans, inplace=True)

from linearmodels.iv import IV2SLS
import statsmodels.api as sm

# Définir la variable dépendante, les variables exogènes et endogènes, et les instruments
Y = filtered_data['delta_CASH']  # Variable dépendante en première différence
X = sm.add_constant(filtered_data[[f'delta_{var}' for var in control_vars]])  # Variables exogènes en première différence
endog = filtered_data['delta_CASH']  # Variable endogène à instrumentaliser
instruments = filtered_data[['CASH_lag1'] + [f'{var}_lag1' for var in control_vars]]  # Instruments

# Régression IV
model = IV2SLS(dependent=Y, exog=X, endog=endog, instruments=instruments).fit(cov_type='unadjusted')

# Affichage des résultats
print(model.summary)

import numpy as np
import pandas as pd
from linearmodels import PanelOLS, RandomEffects
from linearmodels.panel import compare

# Ensure 'filtered_data' is your DataFrame name

# Check if 'year' column exists and is of the correct type
if 'year' in filtered_data.columns:
    filtered_data['year'] = filtered_data['year'].astype(int)
else:
    raise KeyError("Column 'year' does not exist in DataFrame.")

# Safely transform 'totalassets' and 'totalDebt'
filtered_data['log_TotalAssets'] = np.log(filtered_data['totalassets'].clip(lower=1))
filtered_data['log_TotalDebt'] = np.log(filtered_data['totalDebt'].clip(lower=1))

# Calculate Efficiency and R&D Intensity
filtered_data['Efficiency'] = filtered_data['totalexpenses'] / filtered_data['totalrevenues'].clip(lower=1)
filtered_data['RnDIntensity'] = filtered_data['wage_salaries'] / filtered_data['totalrevenues'].clip(lower=1)

# Check and set the MultiIndex for panel data
if not isinstance(filtered_data.index, pd.MultiIndex):
    if 'siren' in filtered_data.columns and 'year' in filtered_data.columns:
        filtered_data = filtered_data.set_index(['siren', 'year'])
    else:
        raise KeyError("One or both of the 'siren' and 'year' columns do not exist, or the index is already set.")

# Define model formulas
formulas = {
    'Profitability': 'roa ~ log_TotalAssets + log_TotalDebt + InterestR',
    'Operational Efficiency': 'netincome ~ Efficiency + wage_salaries',
    'Financial Structure': 'financialdebtequity ~ sharecapital + totalDebt',
    'Investment in Intangibles': 'NetIntangible ~ RnDIntensity + totalrevenues',
    'Liquidity Management': 'currentratio ~ currentas + creditors'
}

# Loop through formulas to fit FE and RE models, then compare them
for title, formula in formulas.items():
    # Fixed Effects Model
    mod_FE = PanelOLS.from_formula(f'{formula} + EntityEffects', data=filtered_data)
    res_FE = mod_FE.fit(cov_type='clustered', cluster_entity=True)

    # Random Effects Model
    mod_RE = RandomEffects.from_formula(formula, data=filtered_data)
    res_RE = mod_RE.fit()

    # Compare models and perform Hausman test
    print(f"Results for {title}:")
    comparison = compare({"Fixed Effects": res_FE, "Random Effects": res_RE})
    print(comparison.summary)
```
